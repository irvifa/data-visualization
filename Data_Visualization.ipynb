```python
from datapackage import Package

package = Package('https://datahub.io/sports-data/italian-serie-a/datapackage.json')
print(package.resource_names)
```

```python
seasons_data = [resource.descriptor["name"] for resource in package.resources if resource.descriptor['datahub']['type'] == 'derived/csv']
print(seasons_data)
```

```python
from datapackage import Resource
import pandas as pd

for resource in seasons_data:
  df = pd.DataFrame.from_dict(package.get_resource(resource).read(keyed=True))
  df = df.assign(Season=resource.split('_')[0])
  print(resource)
  print(df.columns)
  print(df.head())
```

```python
combined_csv = pd.concat(pd.DataFrame.from_dict(pd.DataFrame.from_dict(package.get_resource(resource).read(keyed=True))).assign(Season=resource.split('_')[0]) for resource in seasons_data)
combined_csv
```

```python
combined_csv.to_csv("data.csv", index=False, mode='w+', encoding='utf-8-sig')
```

```python
print(combined_csv.columns)
```

```python
filtered_rows = ['Season', 'Date', 'AwayTeam', 'HomeTeam', 'AS', 'HS', 'AST', 'HST', 'HTAG', 'HTHG', 'HTR', 'FTAG', 'FTHG', 'FTR', 'AC', 'HC', 'AF', 'HF', 'AY', 'HY', 'AR', 'HR']
```

```python
filtered_out_ds = combined_csv[filtered_rows]
filtered_out_ds.head()
```

```python
# Half Time Result (H=Home Win, D=Draw, A=Away Win) ║ H=3, D=1, A=0

htr_score_mapping = {"H": 3, "D": 1, "A": 0}
filtered_out_ds['HTR'] = filtered_out_ds['HTR'].map(htr_score_mapping)
```

```python
# Full Time Result (H=Home Win, D=Draw, A=Away Win) ║ H=3, D=1, A=0

ftr_score_mapping = {"H": 3, "D": 1, "A": 0}
filtered_out_ds['FTR'] = filtered_out_ds['FTR'].map(ftr_score_mapping)
```
